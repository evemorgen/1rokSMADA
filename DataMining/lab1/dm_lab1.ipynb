```r
# 2 Load the cars data set
data(cars)
```

```r
# 2.1 Create a copy of the data, say, cars2.
cars2 <- cars
```

```r
# 2.2 Modify cars2: transform speed units into m/s (meters per second), and distance units into m (meters).
cars2
```

```r
cars2$speed = cars2$speed * 1.609344
cars2$dist = cars2$dist * 0.3048
```

```r
cars2
```

```r
# 2.3 Create a scatterplot for the variables speed and dist.
plot(cars2$speed ~ cars2$dist)
```

```r
# 2.4 Use linear regression to find the best line to fit speed and dist variables.
plot(cars2$speed ~ cars2$dist)
abline(lm(cars2$speed ~ cars2$dist))
```

```r
# 2.5 Extend the data-frame with the third column (dist2), which contains the speed in km/h (kilometers per hour)
cars2$dist2 = cars2$dist * 3.6
```
